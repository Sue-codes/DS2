### Calculate the mean of each sample's 'AGE'.

```python
import numpy as np

print(" mean of sample1 ", np.mean(sample1['AGE']),
      "\n mean of sample2 ", np.mean(sample2['AGE']),
      "\n mean of sample3 ", np.mean(sample3['AGE']),
      "\n mean of sample4 ", np.mean(sample4['AGE']),
      "\n mean of sample5 ", np.mean(sample5['AGE']),
      "\n mean of sample6 ", np.mean(sample6['AGE'])) 
```

### Consider the whole dataset's 'AGE' as the population, calculate the mean of an sample of 10 people, as well as the mean of the full population. Are they the same or different, why?

```python


P_Age=boston_df['AGE']

Sample_P_Age=P_Age.sample(n=10)

print(np.mean(Sample_P_Age))
print(np.mean(P_Age))

## There's a small difference because of the way the data was sampled
```
